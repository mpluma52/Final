```python
import numpy as np
import matplotlib.pyplot as plt

data=np.loadtxt("olympic100m_men.txt",delimiter=",").T
x=data[0]
y=data[1]
plt.scatter(x,y)

N=len(x)

m=(sum(x*y)-sum(x)*sum(y)*1/N)/(sum(x**2)-((sum(x))**2)/N)
b=(sum(y)-m*sum(x))/N

print("Pendiente: ")
print(m)

print("Ordenada al origen: ")
print(b)
```
